```python
from Functions_24_Buda import *

# select the first 2 letters of my surname
letters = ['B', 'U']
```

```python
C = 0.1
gamma = 0.1

X_train, y_train, X_test, y_test = SVM_data_loading(*letters)
model = get_model_MVPSMO(C = C, gamma = gamma, kernel_type = 'gaussian')
model.fit(X_train, y_train)
```

```python
print(f"This gaussian kSVM was trained with C = {C} and γ = {gamma} with performances:")
print(f"Training accuracy (on the complete training dataset)= {model.evaluate(X_train, y_train)}")
print(f"Test accuracy = {model.evaluate(X_test, y_test)}")
print(f"Time necessary for the optimization = {model.exec_time} seconds")
print(f"Number of optimization iterations = {model.n_iter}")
print(f"Final difference between m(λ) and M(λ) = {model.final_violation}")
print(f"Final value of the objective function of the dual = {model.final_dual_obj}")
plot_confusion_matrices(model, X_train, y_train, X_test, y_test, letters)
```
